```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sms
import matplotlib.pyplot as plt
np.random.seed(2)
```

# Taking Credit card data


```python
data = pd.read_csv('creditcard.csv')
```

#  Data Exploration


```python
data.head()
```

```python
data.tail()
```

# Pre-processing

```python
from sklearn.preprocessing import StandardScaler
data['normalizedAmount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1,1))
data = data.drop(['Amount'],axis=1)
```

```python
data.head()
```

```python
data = data.drop(['Time'],axis=1)
data.head()
```

```python
X = data.iloc[:,data.columns !='Class']
y = data.iloc[:,data.columns =='Class']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state= 0)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
```

```python
random_forest.fit(X_train,y_train.values.ravel())
```

```python
y_pred = random_forest.predict(X_test)
```

```python
random_forest.score(X_test,y_test)
```

```python
import pickle
result_rf = random_forest.score(X_test,y_test)

with open('rf.pk1','wb') as my_file_obj:
    pickle.dump(result_rf,my_file_obj)
print('File stored Successfully')
```

```python
cnf_matrix = confusion_matrix(y_test,y_pred)
```

```python
labels= [0,1]
sms.heatmap(cnf_matrix,annot=True, cmap='YlGnBu',fmt = ".3f",xticklabels= labels, yticklabels= labels)
plt.show()
```
